##### <br>TD-IDF pairwise similarity <br>

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vect = TfidfVectorizer(min_df=1, stop_words="english")                                                                                                                                                                                                   
tfidf = vect.fit_transform(BookLists)                                                                                                                                                                                                                       
pairwise_similarity = tfidf * tfidf.T 
```

```python
pairwise_similarity
```

```python
import pandas as pd
import scipy.sparse

simillarity_array = scipy.sparse.csr_matrix.toarray(pairwise_similarity)
pd.DataFrame.sparse.from_spmatrix(pairwise_similarity)
```

```python
import numpy as np
most_similar = []
for idx, i in enumerate(simillarity_array):
    list_init = []
    list_idx = []
    for j in i:
        list_init.append(j)
    list_sort = list_init
    list_sort.sort()
    maxVal = list_sort[-2]
    idy = list_init.index(maxVal)
    most_similar.append([idx,idy,maxVal])
#     for id, j in enumerate(i):
#         if j == 1:
#             pass
#         elif j < 1:
#             list_init.append(j)
#             list_idx.append(id)
#     max_val = max(list_init)
#     index_max_val_list = list_init.index(max_val)
#     index_max_val = list_idx(index_max_val_list)

#     most_similar.append([idx,index_max_val,max_val])
most_similar    
```
